$x$ is actual resistance 

$v_i$ is a measurement noise

Noise is independent and identically distributed or IID

The 'best' estimate of resistacne is the one that minimizes the sum of squared error


That is the best estimate of $x$ is one which **minimizes the squared error criterion**, also sometimes called **the squared error cost function** or **loss function**. To minimize the squared error criterion, we'll rewrite our errors in matrix notation.

 $H$ has the dimensions $m$ by $n$, where 
 
 $m$ is the number of measurements and 
 
 $n$ is the number of unknowns or parameters that we wish to estimate.

$2x^TH^TH = 2y^TH$

$x^TH^TH = y^TH$

$(x^TH^TH)^T = (y^TH)^T$

$(H^TH)^Tx = H^Ty$

$H^THx = H^Ty$

$x =(H^TH)^{-1} H^Ty$

$ -2y^TH + 2x^TH^TH = 0$

$ 2(-y^TH + x^TH^TH) = 0$

$ -y^TH + x^TH^TH = 0$

$ (-y^T + x^TH^T)H = 0$

$ -y^T + x^TH^T = 0$

$  x^TH^T = y^T$

$  Hx = y$

$  x = H^{-1}y$

$H^{-1} = (H^TH)^{-1} H^T$

$(H^TH) H^{-1} =  H^T$

$H^THH^{-1} =  H^T$

$H^T =  H^T$
